```python
arenas = pd.read_csv("Arenas.csv")
arenas[:]

type(arenas['COMMUNITY'][3])

for i in range(len(arenas)):
    arenas['ADDRESS'][i] = ('Address: ' + arenas['ADDRESS'][i] + '\nCommunity: ' + arenas['COMMUNITY'][i] + '\nPhone: ' + arenas['PHONE'][i])

arenas.rename(columns={'ADDRESS':'DESCRIPTION'}, inplace=True)

arenas = arenas[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]

arenas.to_csv("Arenas_NEW.csv")
```

```python
beaches = pd.read_csv("Beaches.csv")
beaches.head()

beaches.rename(columns={'WATER_SOURCE':'DESCRIPTION'}, inplace=True)

beaches = beaches[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]

print(beaches.head())
```

```python
campgrounds = pd.read_csv("Campgrounds.csv")
print(campgrounds.head())

for i in range(len(campgrounds)):
    campgrounds['LOCATION'][i] = ('Address: ' + campgrounds['LOCATION'][i] + '\nURL: ' + campgrounds['URL'][i])

campgrounds.rename(columns={'LOCATION':'DESCRIPTION'}, inplace=True)

campgrounds = campgrounds[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]

print(campgrounds.head())
```

```python
falls = pd.read_csv("City_Waterfalls.csv")
print(falls.head())

for i in range(len(falls)):
    falls['CLUSTER_AREA'][i] = ('Area: ' + falls['CLUSTER_AREA'][i] + '\nAlternate Name: ' + str(falls['ALTERNATE_NAME'][i]) + '\nHeight: ' + str(falls['HEIGHT_IN_M'][i]) + 'm' + '\nWidth: ' + str(falls['WIDTH_IN_M'][i]) + 'm')

falls.rename(columns={'CLUSTER_AREA':'DESCRIPTION'}, inplace=True)

falls = falls[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]
print(falls.head())
```

```python
edu_institutions = pd.read_csv("Educational_Institutions.csv")
print(edu_institutions.head())

a = ('Address: ' + edu_institutions['ADDRESS'][2] + '\nCommunity: ' + edu_institutions['COMMUNITY'][2] + '\nCategory: ' + edu_institutions['CATEGORY'][2] + '\nSchool Board: ' + edu_institutions['SCHOOL_BOARD'][2])
print(a)
for i in range(len(edu_institutions)):
    edu_institutions['ADDRESS'][i] = ('Address: ' + edu_institutions['ADDRESS'][i] + '\nCommunity: ' + edu_institutions['COMMUNITY'][i] + '\nCategory: ' + edu_institutions['CATEGORY'][i] + '\nSchool Board: ' + str(edu_institutions['SCHOOL_BOARD'][i]))

edu_institutions.rename(columns={'ADDRESS':'DESCRIPTION'}, inplace=True)

edu_institutions = edu_institutions[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]
print(edu_institutions.head())
```

```python
libs = pd.read_csv("Libraries.csv")
libs.head()

for i in range(len(libs)):
    libs['HOLDINGS'][i] = ('Holdings: ' + str(libs['HOLDINGS'][i]) + '\nCirculation: ' + str(libs['CIRCULATION'][i]))

libs.rename(columns={'HOLDINGS':'DESCRIPTION'}, inplace=True)

libs = libs[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]

print(libs.head())
```

```python
museums_and_galleries = pd.read_csv("Museums_and_Galleries.csv")
museums_and_galleries.head()

for i in range(len(museums_and_galleries)):
    museums_and_galleries['OBJECTID'][i] = ('N/A')

museums_and_galleries.rename(columns={'OBJECTID':'DESCRIPTION'}, inplace=True)

museums_and_galleries = museums_and_galleries[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]
print(museums_and_galleries.head())
```

```python
park_amenities = pd.read_csv("Park_Amenities.csv")
print(park_amenities.head())

for i in range(len(park_amenities)):
    park_amenities['TYPE'][i] = (str(park_amenities['TYPE'][i]) + 'ball')
    park_amenities['REC_AREA'][i] = ('Recreational Area: ' + str(park_amenities['REC_AREA'][i]) + 'Ownership: ' + str(park_amenities['OWNERSHIP'][i]))

park_amenities.rename(columns={'TYPE':'NAME'}, inplace=True)
park_amenities.rename(columns={'REC_AREA':'DESCRIPTION'}, inplace=True)

park_amenities = park_amenities[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]
print(park_amenities.head())
```

```python
worship = pd.read_csv("Places_of_Worship.csv")
print(worship.head())

for i in range(len(worship)):
    worship['DENOMINATION'][i] = ('Denomination: ' + worship['DENOMINATION'][i] + '\nAddress: ' + worship['ADDRESS'][i] + '\nCommunity: ' + worship['COMMUNITY'][i])

worship.rename(columns={'DENOMINATION':'DESCRIPTION'}, inplace=True)

worship = worship[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]
print(worship.head())
```

```python
public_art = pd.read_csv("Public_Art_and_Monuments.csv")
print(public_art.head())

for i in range(len(public_art)):
    public_art['DESCRIPTION'][i] = ('Address: ' + public_art['STREET'][i] + '\nDescription: ' + public_art['DESCRIPTION'][i])

public_art.rename(columns={'ARTWORK_TITLE':'NAME'}, inplace=True)

public_art = public_art[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]

print(public_art.head())
```

```python
rec_centres = pd.read_csv("Recreation_and_Community_Centres.csv")
print(rec_centres.head())

for i in range(len(rec_centres)):
    rec_centres['OBJECTID'][i] = ('N/A')

rec_centres.rename(columns={'OBJECTID':'DESCRIPTION'}, inplace=True)

rec_centres = rec_centres[['LATITUDE','LONGITUDE','NAME','DESCRIPTION']]

print(rec_centres.head())
```
